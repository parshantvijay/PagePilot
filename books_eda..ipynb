```python
#both books.csv and ratings.csv have ISBN common.
#merging ratings with books
merged_ratings_books = ratings.merge(books,on='ISBN')
merged_ratings_books 
```

```python
book_rating_count = merged_ratings_books.groupby('Book-Title').count()['Book-Rating'].reset_index()
book_rating_count.rename(columns={'Book-Rating':'Rating-Count'}, inplace =True)
book_rating_count
```

```python
avg_rating_count = merged_ratings_books.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating_count.rename(columns={'Book-Rating':'Average-Rating'}, inplace =True)
avg_rating_count
```

```python
popular_books = book_rating_count.merge(avg_rating_count,on='Book-Title')
popular_books
```

```python
#keeping only those books in popular whose rating_count is >=250
popular_books = popular_books[popular_books['Rating-Count']>=250].sort_values('Average-Rating',ascending=False)
popular_books
```

```python
popular_books = popular_books.merge(books, on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Year-Of-Publication','Image-URL-L','Rating-Count','Average-Rating']]
popular_books
# [['Book-Title','Book-Author','Year-Of-Publication','Image-URL-M','Rating-Count','Average-Rating']]
```

## Colaborative Filtering Item Based Recommender System

```python
temp = merged_ratings_books.groupby('User-ID').count()['Book-Rating']>200
high_rating_users = temp[temp].index
high_rating_users
```

```python
filtered_rating = merged_ratings_books[merged_ratings_books['User-ID'].isin(high_rating_users)]
filtered_rating
```

```python
temp2 = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
popular_books_by_highratedusers= temp2[temp2].index
popular_books_by_highratedusers
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(popular_books_by_highratedusers)]
```

```python
final_ratings.duplicated().sum()
```

```python
#no duplicates found in the final and cleaned version of book ratings. 
#now we will create a pivot table showing book names as indexes, column as User-Ids and the values will be the rating given to each book by each user

thepivottable= final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
thepivottable
```

```python
thepivottable.fillna(0,inplace=True)
thepivottable
```

```python
from sklearn.metrics.pairwise import cosine_similarity
books_similarity_scores = cosine_similarity(thepivottable)
books_similarity_scores
```

```python
np.where(thepivottable.index =='Zoya')[0][0]
# basically it finds the index of the book in the thepivottable
sorted(list(enumerate(books_similarity_scores[704])), key=lambda x:x[1], reverse=True)[1:6]
#most similar items are now at top
# thepivottable.index[0]
```

```python
def recommend(book_name):
    index = np.where(thepivottable.index == book_name)[0][0]  
    distance = books_similarity_scores[index]
    similar_books = sorted(list(enumerate(books_similarity_scores[index])), key=lambda x:x[1], reverse=True)[1:11]
    total_kitaab = []
    for i in similar_books:
        kitaab = []
        temp3 = books[books['Book-Title'] == thepivottable.index[i[0]]]
        kitaab.extend(list(temp3.drop_duplicates('Book-Title')['Book-Title'].values))
        kitaab.extend(list(temp3.drop_duplicates('Book-Title')['Book-Author'].values))
        kitaab.extend(list(temp3.drop_duplicates('Book-Title')['IMG-URL-L'].values))

        total_kitaab.append(kitaab)

    return total_kitaab
```
